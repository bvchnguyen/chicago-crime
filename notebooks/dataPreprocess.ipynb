```python
# Import all the libraries needed for the project
# Can be removed if not needed later on

import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
import math as math
```

```python
def data_preprocessing(fileName):

    # Read the data from the csv file
    # prepocessing the data
    # See project proposal for details to consider

    df = pd.read_csv(fileName)
   
    # Drop the rows that have duplicate values
    df = df.drop_duplicates()
    # Drop the rows that have missing values / NaN values
    df = df.dropna()

    # Drop the columns that are not needed, add more if needed
    df.drop("IUCR", axis=1, inplace=True)
    df.drop("FBI Code", axis=1, inplace=True)
    df.drop("X Coordinate", axis=1, inplace=True)
    df.drop("Y Coordinate", axis=1, inplace=True)

    # Split the date and time into two columns
    df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y %I:%M:%S %p')
    df['Time of crime'] = df['Date'].dt.strftime('%H:%M:%S')
    df['Date'] = df['Date'].dt.date

    # Rearrange the columns to have date and time of crime next to each other
    df = df[['ID', 'Case Number', 'Date', 'Time of crime', 
             'Block', 'Primary Type', 'Description', 
             'Location Description', 'Arrest', 'Domestic', 
             'Beat', 'District', 'Ward', 'Community Area', 
             'Year', 'Updated On', 'Latitude', 'Longitude', 'Location']]


    df.to_csv('../data/ChicagoCrime_cleaned.csv', index=False)

    return df
```

```python
cleaned_Data = data_preprocessing('../data/Crimes_-_2001_to_Present.csv')

```
